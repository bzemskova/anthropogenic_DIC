```python
import numpy as np
import os
from scipy.stats import linregress
import matplotlib.pyplot as plt
import matplotlib as mpl
import warnings
warnings.filterwarnings('ignore')
import h5py
```

#### 1) Open $C^*$ values (in mmol/m$^3$) for all years (1992-2022) and store them as a stacked array.

For each year, $C^*(x,y,z)$ is on a regular (lat, lon, depth) grid with $(180, 360, 50)$ grid points.

```python
data_folder = "../raw_Cstar_data/"
years = list(range(1992, 2023))  #31 years: 1992 to 2022
cstar_stack = []

#Load each year's file and append to list
for y in years:
    path = os.path.join(data_folder, f"cstar_interp_{y}.h5")
    with h5py.File(path, 'r') as f:
        cstar = f['Cstar'][:] # Use [:] to load data into memory
    cstar_stack.append(cstar)

# load latitude and depth arrays
with h5py.File(path, 'r') as f:
    lat = f['latitude'][:] 
    depth = f['depth'][:] 

#Stack
cstar_all = np.stack(cstar_stack, axis=-1)         
print('Shape of stacked array (lat, lon, depth, years): ', cstar_all.shape)
```

```python
# Load \Delta z (vertical grid spacing)
drf_path = "../processed_data/DRF.data" 
drf = np.fromfile(drf_path, dtype='>f4')[:50]

# Load ocean basin codes (1: Atlantic, 2: Pacific, 3: Indian, 4: Southern Ocean, 5: Arctic)
with h5py.File('../processed_data/basin_codes.h5', 'r') as f:
    basin_code = f['basin_code'][:]

# Load area for each grid point
with h5py.File('../processed_data/area_volume.h5', 'r') as f:
    area = f['Area'][:]
```

#### 2) Compute zonally-averaged $\Delta C^*$ for each ocean basin (Atlantic, Pacific, Indian, Southern, Arctic)

For each grid point $(x,y,z)$, the annual timeseries $C^*(x,y,z,t)$ is extracted over a specified time period (a particular decade or the full time period of analysis). $\Delta C^*$ is computed using linear regression over that specified time period.

```python
def compute_basin_zonal_trend(data, basin_mask, basin_id, years, start, end, lat):
    # only needs 2D basin code information
    if basin_mask.ndim == 3:
        basin_mask = basin_mask[:, :, 0]

    # specify year ranges
    idx = [i for i, y in enumerate(years) if start <= y <= end]
    x = np.array(years)[idx]
    dC = np.full((180, 360, 50), np.nan) # rate of change of C* at each grid point over the specified time period
    dC_int = np.full((180, 50), np.nan) # rate of change of C* zonally integrated

    # Compute linear rate of change at each grid point over the specified time period
    for i in range(180):
        for j in range(360):
            for k in range(50):
                y = data[i, j, k, idx]
                if np.isnan(y).all(): # ignore all NaNs (land)
                    continue
                dC[i, j, k] = linregress(x, y).slope

        for k in range(50):
            basin_j = (basin_mask[i, :] == basin_id) # Extract points for a particular ocean basin
            if not np.any(basin_j):
                continue
            temp = dC[i, :, k] # extract all values at a particular latitude and depth
            masked = temp[basin_j] # extract values within a particular basin
            land_mask = masked / masked # land mask (for regions that are not part of that basin)
            if np.isnan(masked).all(): 
                continue
            total_area = np.nansum(land_mask * area[i, 0, 0]) # total zonal area
            # Zonally-averaged \Delta C^* in mol/m^3/yr
            dC_int[i, k] = np.nansum(masked * area[i, 0, 0]) / total_area / 1000

    return dC_int
```

```python
basin_ids = [1, 2, 3, 4, 5]
basin_names = ['Atlantic', 'Pacific', 'Indian', 'Southern', 'Arctic']

# Atlantic: different time periods
deltaC_atl_1993_2002 = compute_basin_zonal_trend(cstar_all, basin_code, 1, years, 1993, 2002, lat)
deltaC_atl_2003_2012 = compute_basin_zonal_trend(cstar_all, basin_code, 1, years, 2003, 2012, lat)
deltaC_atl_2013_2022 = compute_basin_zonal_trend(cstar_all, basin_code, 1, years, 2013, 2022, lat)
deltaC_atl_1993_2022 = compute_basin_zonal_trend(cstar_all, basin_code, 1, years, 1993, 2022, lat)

# Pacific: different time periods
deltaC_pac_1993_2002 = compute_basin_zonal_trend(cstar_all, basin_code, 2, years, 1993, 2002, lat)
deltaC_pac_2003_2012 = compute_basin_zonal_trend(cstar_all, basin_code, 2, years, 2003, 2012, lat)
deltaC_pac_2013_2022 = compute_basin_zonal_trend(cstar_all, basin_code, 2, years, 2013, 2022, lat)
deltaC_pac_1993_2022 = compute_basin_zonal_trend(cstar_all, basin_code, 2, years, 1993, 2022, lat)

# Indian: different time periods
deltaC_ind_1993_2002 = compute_basin_zonal_trend(cstar_all, basin_code, 3, years, 1993, 2002, lat)
deltaC_ind_2003_2012 = compute_basin_zonal_trend(cstar_all, basin_code, 3, years, 2003, 2012, lat)
deltaC_ind_2013_2022 = compute_basin_zonal_trend(cstar_all, basin_code, 3, years, 2013, 2022, lat)
deltaC_ind_1993_2022 = compute_basin_zonal_trend(cstar_all, basin_code, 3, years, 1993, 2022, lat)

# Southern: different time periods
deltaC_so_1993_2002 = compute_basin_zonal_trend(cstar_all, basin_code, 4, years, 1993, 2002, lat)
deltaC_so_2003_2012 = compute_basin_zonal_trend(cstar_all, basin_code, 4, years, 2003, 2012, lat)
deltaC_so_2013_2022 = compute_basin_zonal_trend(cstar_all, basin_code, 4, years, 2013, 2022, lat)
deltaC_so_1993_2022 = compute_basin_zonal_trend(cstar_all, basin_code, 4, years, 1993, 2022, lat)

# Arctic: different time periods
deltaC_arc_1993_2002 = compute_basin_zonal_trend(cstar_all, basin_code, 5, years, 1993, 2002, lat)
deltaC_arc_2003_2012 = compute_basin_zonal_trend(cstar_all, basin_code, 5, years, 2003, 2012, lat)
deltaC_arc_2013_2022 = compute_basin_zonal_trend(cstar_all, basin_code, 5, years, 2013, 2022, lat)
deltaC_arc_1993_2022 = compute_basin_zonal_trend(cstar_all, basin_code, 5, years, 1993, 2022, lat)
```

#### 3) Reproduce manuscript figures: 

- rate of change of $C^*$ over the entire period (1992-2022)
- rate of change of $C^*$ for each decadal period (1993-2002, 2003-2012, 2013-2022)

```python
fig, axs = plt.subplots(5,1, sharey=True, sharex=True)

cmap = plt.get_cmap("seismic").copy()
cmap.set_bad(color="lightgray")
vmax = 5e-3

axs[0].pcolormesh(lat, -depth/1e3, np.ma.masked_invalid(deltaC_atl_1993_2022.T), vmin = -vmax, vmax = vmax, cmap=cmap)
axs[0].set(ylabel = 'Depth \n  (km)', title = '(a) Atlantic')
for item in ([axs[0].title, axs[0].xaxis.label, axs[0].yaxis.label]):
    item.set_fontsize(12)

axs[1].pcolormesh(lat, -depth/1e3, np.ma.masked_invalid(deltaC_pac_1993_2022.T), vmin = -vmax, vmax = vmax, cmap=cmap)
axs[1].set(ylabel = 'Depth \n  (km)',title = '(b) Pacific')
for item in ([axs[1].title, axs[1].xaxis.label, axs[1].yaxis.label]):
    item.set_fontsize(12)

axs[2].pcolormesh(lat, -depth/1e3, np.ma.masked_invalid(deltaC_ind_1993_2022.T), vmin = -vmax, vmax = vmax, cmap=cmap)
axs[2].set(ylabel = 'Depth \n  (km)', title = '(c) Indian')
for item in ([axs[2].title, axs[2].xaxis.label, axs[2].yaxis.label]):
    item.set_fontsize(12)

axs[3].pcolormesh(lat, -depth/1e3, np.ma.masked_invalid(deltaC_so_1993_2022.T), vmin = -vmax, vmax = vmax, cmap=cmap)
axs[3].set(ylabel = 'Depth \n (km)', title = '(d) Southern')
for item in ([axs[3].title, axs[3].xaxis.label, axs[3].yaxis.label]):
    item.set_fontsize(12)

im = axs[4].pcolormesh(lat, -depth/1e3, np.ma.masked_invalid(deltaC_arc_1993_2022.T), vmin = -vmax, vmax = vmax, cmap=cmap)
axs[4].set(ylabel = 'Depth \n  (km)', xlabel = 'Latitude', title = '(e) Arctic')
for item in ([axs[4].title, axs[4].xaxis.label, axs[4].yaxis.label]):
    item.set_fontsize(12)

fig.subplots_adjust(bottom=0.25, hspace=0.35)# hspace=0.2)
cbar_ax = fig.add_axes([0.2, 0.15, 0.6, 0.015])
cb = fig.colorbar(im, cax=cbar_ax, orientation="horizontal", pad=0.2)
cb.ax.tick_params(labelsize=12)
cb.set_label(r'$\widetilde{\Delta C^*}$ (mol/m$^3$/yr)', fontsize=14)

fig.set_size_inches(6,9)
fig.savefig("../figures/zon_int_deltaCstar.png", dpi=600, bbox_inches='tight')
```

```python
str_years = ['_1993_2002','_2003_2012', '_2013_2022']
str_basins = ['atl', 'pac', 'ind', 'so', 'arc']
names = ['Atlantic:', 'Pacific:', 'Indian:', 'Southern:', 'Arctic:']
yrs = ['1993-2002', '2003-2012', '2013-2022']
title_list = ['(a)', '(b)', '(c)', '(d)', '(e)', '(f)', '(g)', '(h)', '(i)', '(j)', '(k)',
             '(l)', '(m)', '(n)', '(o)', '(p)', '(q)', '(r)', '(s)', '(t)', '(u)', '(v)', '(w)', '(x)', '(y)', '(z)']
file_names = ['zonal_dCstar_atlantic.h5', 'zonal_dCstar_pacific.h5', 
              'zonal_dCstar_indian.h5', 'zonal_dCstar_southern.h5', 'zonal_dCstar_arctic.h5']

cmap = plt.get_cmap("seismic").copy()
cmap.set_bad(color="lightgray")
vmax = 5e-3

rows, cols = 5, 3
fig, axs = plt.subplots(rows,cols, sharey=True, sharex=True)

ind = 0
for indr in range(rows):
    for indc in range(cols):
        filename = '../processed_data/'+file_names[indr]
        varname = 'dCstar' +  str_years[indc]
        with h5py.File(filename, 'r') as f:
            var = f[varname][:]
        im = axs[indr, indc].pcolormesh(lat, -depth/1e3, np.ma.masked_invalid(var.T), vmin = -vmax, vmax = vmax, cmap=cmap)
        title = title_list[ind] + ' ' + names[indr] + ' ' + yrs[indc]
        axs[indr, indc].set(title = title)
        if indr == rows-1:
            axs[indr, indc].set(xlabel = 'Latitude')
        if indc == 0:
            axs[indr, indc].set(ylabel = 'Depth (km)')

        for item in ([axs[indr, indc].title, axs[indr, indc].xaxis.label, axs[indr, indc].yaxis.label]):
            item.set_fontsize(14)
        ind += 1

fig.subplots_adjust(bottom=0.25, wspace=0.05)# hspace=0.2)
cbar_ax = fig.add_axes([0.2, 0.15, 0.6, 0.015])
cb = fig.colorbar(im, cax=cbar_ax, orientation="horizontal", pad=0.2)
cb.ax.tick_params(labelsize=12)
cb.set_label(r'$\widetilde{\Delta C^*}$ (mol/m$^3$/yr)', fontsize=14)

fig.set_size_inches(16,16)
fig.savefig("../figures/zon_int_deltaCstar_decadal.png", dpi=600, bbox_inches='tight')
```
